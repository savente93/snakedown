## SVG

```python
svg = """<svg height="100" width="100" xmlns="http://www.w3.org/2000/svg"><circle cx="50" cy="50" r="40" stroke="black"></circle></svg>"""

class SVG:
    def _repr_svg_(self):
        return svg

display(SVG())
```
